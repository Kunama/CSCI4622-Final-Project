```python
import pandas as pd
import sklearn as sk
import numpy as np
import matplotlib.pyplot as plt
```

```python
names = ["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", "hours-per-week", "native-country", "income"]
df = pd.read_csv("./data/adult.data", names=names, skipinitialspace=True)
```

```python
df.info()
```

```python
#count and replace the ? entries
for column in df.columns:
    print(column + ": {}".format((df[column] == '?').sum()))
df[df == '?'] = np.nan
```

```python
#drop rows with missing values
print("old len: {}".format(df.shape[0]))
df = df.dropna()
print("new len: {}".format(df.shape[0]))
```

```python
#make the income numerical
df["income"] = df["income"].map({"<=50K":0, ">50K":1})
#drop fnlweight as it is only interesting for the census authorities
df = df.drop(labels="fnlwgt", axis=1)
```

```python
for column in df.columns:
    print(df[column].value_counts())
```

The cells education and education-num are equal, which means we can drop the non numerical education column.

```python
df = df.drop(labels="education", axis=1)
```

```python
#visualization of each column in frequency
for column in df.columns:
    values = df[column].value_counts().sort_index()
    categories = df[column].unique()
    ax = values.plot.bar(title=column, figsize=(12,6))
    ax.set_ylabel("Frequency", fontsize=16)
    plt.show()
```

```python
#drop fnlweight as it is only interesting for the census authorities
#df = df.drop(labels="native-country", axis=1)
```

```python
#split the data for the pca and further processing
from sklearn.model_selection import train_test_split

df_x = df.drop(labels="income", axis=1)
df_y = df["income"]

train_x, test_x, train_y, test_y = train_test_split(df_x, df_y, test_size=2/10, random_state=0)
print(train_x.shape)
print(test_x.shape)
```

```python
#convert the categorical data to numerical
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
features_to_encode = ["workclass", "marital-status", "occupation", "relationship", "race", "sex", "native-country"]
for feature in features_to_encode:
    train_x[feature] = le.fit_transform(train_x[feature])
    test_x[feature] = le.fit_transform(test_x[feature])

train_x.head()
```

```python
#perform pca to see if we can drop even more columns which might be correlated. 
#martial status and relationship seem pretty similar to begin with
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

#use labelencoder to encode the string labels to int


#scales the data to the standard normal, as is required for pca of sklearn
scaler = StandardScaler()
scaler.fit(train_x)
train_x = scaler.transform(train_x)
test_x = scaler.transform(test_x)

pca = PCA(.90)
pca.fit(train_x)

train_x = pca.transform(train_x)
test_x = pca.transform(test_x)
print(train_x.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
def test_depth(range_steps, random_state=0):
    scores = []
    for step in range_steps:
        clf = RandomForestClassifier(step, random_state=random_state)
        clf.fit(train_x, train_y)
        scores.append(clf.score(test_x, test_y))
    return scores
```

```python
range_steps_depth = np.arange(1, 100, 1)
scores_depth = test_depth(range_steps_depth)
```

```python
fig, ax = fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,6))
plt.plot(range_steps_depth, scores_depth)
ax.set_title("Accuracy with varying max tree depth", fontsize=20)
ax.set_ylabel("Accuracy", fontsize=16)
ax.set_xlabel("Max depth of the tree", fontsize=16)
plt.show()
```

```python
def test_min_sample_split(range_steps, random_state=0):
    scores = []
    for step in range_steps:
        clf = RandomForestClassifier(50,min_samples_split=step, random_state=random_state)
        clf.fit(train_x, train_y)
        scores.append(clf.score(test_x, test_y))
    return scores
```

```python
range_steps_split = np.arange(2, 50, 1)
scores_split = test_min_sample_split(range_steps_split)
```

```python
fig, ax = fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,6))
plt.plot(range_steps_split, scores_split)
ax.set_title("Accuracy with varying max min sample split", fontsize=20)
ax.set_ylabel("Accuracy", fontsize=16)
ax.set_xlabel("min sample split", fontsize=16)
plt.show()
```

```python
def test_max_features(range_steps, random_state=0):
    scores = []
    for step in range_steps:
        clf = RandomForestClassifier(50, max_features=step, random_state=random_state)
        clf.fit(train_x, train_y)
        scores.append(clf.score(test_x, test_y))
    return scores
```

```python
range_max_features = np.arange(1, 11, 1)
scores_max_features = test_max_features(range_max_features)
```

```python
fig, ax = fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,6))
plt.plot(range_max_features, scores_max_features)
ax.set_title("Accuracy with varying max features", fontsize=20)
ax.set_ylabel("Accuracy", fontsize=16)
ax.set_xlabel("max features", fontsize=16)
plt.show()
```

```python
def test_max_leaf_nodes(range_steps, random_state=0):
    scores = []
    for step in range_steps:
        clf = RandomForestClassifier(50, max_leaf_nodes=step, random_state=random_state)
        clf.fit(train_x, train_y)
        scores.append(clf.score(test_x, test_y))
    return scores
```

```python
range_max_nodes = np.arange(2, 50, 1)
scores_max_nodes = test_max_leaf_nodes(range_max_nodes)
```

```python
fig, ax = fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,6))
plt.plot(range_max_nodes, scores_max_nodes)
ax.set_title("Accuracy with varying max leaf nodes", fontsize=20)
ax.set_ylabel("Accuracy", fontsize=16)
ax.set_xlabel("max leaf nodes", fontsize=16)
plt.show()
```

```python
def test_ccp_alpha(range_steps, random_state=0):
    scores = []
    for step in range_steps:
        clf = RandomForestClassifier(50, ccp_alpha=step, random_state=random_state)
        clf.fit(train_x, train_y)
        scores.append(clf.score(test_x, test_y))
    return scores
```

```python
range_ccp_alpha = np.arange(0.0001, 0.02, 0.0002)
scores_ccp_alpha = test_ccp_alpha(range_ccp_alpha)
```

```python
fig, ax = fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,6))
plt.plot(range_ccp_alpha, scores_ccp_alpha)
ax.set_title("Accuracy with varying ccp alpha", fontsize=20)
ax.set_ylabel("Accuracy", fontsize=16)
ax.set_xlabel("ccp alpha", fontsize=16)
plt.show()
```

```python
def test_max_samples(range_steps, random_state=0):
    scores = []
    for step in range_steps:
        clf = RandomForestClassifier(50, max_samples=step, random_state=random_state)
        clf.fit(train_x, train_y)
        scores.append(clf.score(test_x, test_y))
    return scores
```

```python
range_samples = np.arange(20, 2000, 10)
scores_samples = test_max_samples(range_samples)
```

```python
fig, ax = fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,6))
plt.plot(range_samples, scores_samples)
ax.set_title("Accuracy with varying max features", fontsize=20)
ax.set_ylabel("Accuracy", fontsize=16)
ax.set_xlabel("max features", fontsize=16)
plt.show()
```
